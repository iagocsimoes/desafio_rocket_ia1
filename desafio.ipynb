```python
# Dicion√°rio de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
fat = pd.DataFrame(dict_faturamento)
```

```python
fat['data_ref'] = pd.to_datetime(fat['data_ref'])
```

```python
# Media das vendas
fat.valor.mean()
```

```python
# Grafico de Barras Vertical
fat.sort_values('valor').plot.barh(x='data_ref',y='valor')
```

```python
# Grafico de Linha
fat.sort_values('valor').plot.line(x='valor',y='data_ref')
```
